```python
# Define the mapping for the sentiment labels based on your provided classes
class_map = {
    "Anxiety Disorder": 0,
    "Depression": 1,
    "Suicidal Ideation": 2,
    "Stress": 3,
    "Bipolar Disorder": 4,
    "Personality Disorder": 5,
    "Sadness": 6,
    "Happiness": 7
}
# Reverse mapping for id2label
id2label = {v: k for k, v in class_map.items()}

#print(class_map)
print(id2label)
```

*  calculate the class weights (for loss) and sampling weights (for batch balancing).

```python
# Compute class counts
class_counts = train_df['mental_status'].value_counts().to_dict()
total_samples = sum(class_counts.values())
# Compute class weights (Higher weight for less frequent classes)
class_weights = {cls: total_samples / count for cls, count in class_counts.items()}
sample_weights = train_df['mental_status'].map(class_weights).values  # Assign sample weight
# Weighted Sampler
sampler = WeightedRandomSampler(weights=sample_weights, num_samples=len(sample_weights), replacement=True)
```

```python
# Convert weights to tensor for loss function
class_weights_tensor = torch.tensor(list(class_weights.values()), dtype=torch.float).to(device)
# Use Weighted CrossEntropyLoss
criterion = torch.nn.CrossEntropyLoss(weight=class_weights_tensor)
```

```python
print("Class Counts:", class_counts)
print("Class Weights:", class_weights)
```

```python
print(class_weights_tensor)
print(criterion)
```

```python
class SentimentDataset(Dataset):
    def __init__(self, dataframe, tokenizer, return_text_labels=False):
        self.dataset = dataframe
        self.tokenizer = tokenizer
        self.targets = dataframe.mental_status.map(class_map)
        self.return_text_labels = return_text_labels

    def __len__(self):
        return len(self.dataset)

    def clean_text(self, text):
        text = text.lower()
        text = re.sub(r"http\S+|www\S+", "", text)
        text = re.sub(r"@\w+|#\w+", "", text)
        text = re.sub(r"[^a-zA-Z0-9\s,.!?']", "", text)
        text = emoji.demojize(text)
        text = re.sub(r"\s+", " ", text).strip()
        return text

    def __getitem__(self, index):
        text = self.clean_text(str(self.dataset.iloc[index]['statement']))
        label_id = self.targets.iloc[index]
        encoding = self.tokenizer(
            text, truncation=True, padding='max_length', max_length=256, return_tensors='pt'
        )

        label_text = id2label[label_id]  # Convert ID back to text

        return {
            'input_ids': encoding['input_ids'].squeeze(0),
            'attention_mask': encoding['attention_mask'].squeeze(0),
            'labels': torch.tensor(label_id, dtype=torch.long),
            'label_text': label_text
        }

# Creating Dataset
train_dataset = SentimentDataset(train_df, tokenizer, return_text_labels=False)
val_dataset = SentimentDataset(test_df, tokenizer, return_text_labels=False)
# Creating DataLoaders with Weighted Sampler
train_loader = DataLoader(train_dataset, batch_size=32, sampler=sampler)  # Weighted Sampling
val_loader = DataLoader(val_dataset, batch_size=128, shuffle=False)  # Normal validation
```

```python
def contrastive_loss(anchor, positive, negative, margin=0.5):
    pos_sim = cosine_similarity(anchor, positive)  # Similar pairs
    neg_sim = cosine_similarity(anchor, negative)  # Dissimilar pairs
    return torch.mean(torch.clamp(neg_sim - pos_sim + margin, min=0.0))  # Maximize margin
```

Fifth updated training Loop

```python
# Training Loop
# 1️⃣ Setup & Initialization
scaler = torch.cuda.amp.GradScaler()
gradient_accumulation_steps = 4
best_val_loss = float("inf")
patience, patience_counter = 3, 0
train_losses, val_losses, val_accuracies = [], [], []

# 2️⃣ Training Loop (Epochs)
for epoch in tqdm(range(num_epochs), desc="Training Progress"):
    model.train()
    epoch_train_loss = 0.0
# 3️⃣ Training Step (Gradient Accumulation + Mixed Precision)
    for step, batch in enumerate(train_loader):
        batch = {k: v.to(device) for k, v in batch.items() if isinstance(v, torch.Tensor)}

        with torch.cuda.amp.autocast():
            outputs = model(**batch)
            logits = outputs.logits
            loss = criterion(logits, batch['labels'])

            # ✅ Contrastive Learning
            if "embeddings" in outputs:
                contrastive_loss_val = contrastive_loss(outputs.embeddings, batch["labels"])
                loss += 0.1 * contrastive_loss_val  # ✅ Blend Contrastive Loss (λ=0.1)

# 4️⃣ Gradient Accumulation & Optimization
        scaler.scale(loss).backward()
        if (step + 1) % gradient_accumulation_steps == 0:
            scaler.unscale_(optimizer)
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
            scaler.step(optimizer)
            scaler.update()
            optimizer.zero_grad()
            scheduler.step()
        epoch_train_loss += loss.item()
    train_losses.append(epoch_train_loss / len(train_loader))

# 5️⃣ Validation Phase
    model.eval()
    epoch_val_loss, correct, total = 0.0, 0, 0
    with torch.no_grad():
        for batch in val_loader:
            batch = {k: v.to(device) for k, v in batch.items() if isinstance(v, torch.Tensor)}
            outputs = model(**batch)
            loss = criterion(outputs.logits, batch['labels'])
            epoch_val_loss += loss.item()

            preds = torch.argmax(outputs.logits, dim=-1)
            correct += (preds == batch['labels']).sum().item()
            total += batch['labels'].size(0)

    val_losses.append(epoch_val_loss / len(val_loader))
    val_accuracies.append(correct / total)

    print(f"Epoch {epoch}: Train Loss = {train_losses[-1]:.4f} | Val Loss = {val_losses[-1]:.4f} | Val Acc = {val_accuracies[-1]:.4f}")

#6️⃣ Early Stopping & Model Saving
    if val_losses[-1] < best_val_loss:
        best_val_loss = val_losses[-1]
        patience_counter = 0

        # Define save path in Google Drive
        save_path = "/content/drive/MyDrive/MHRoberta_Models"
        # Create directory if it doesn't exist
        os.makedirs(save_path, exist_ok=True)
        model.save_pretrained(f"{save_path}/MHRoberta_epoch_{epoch}")
        tokenizer.save_pretrained(f"{save_path}/MHRoberta_epoch_{epoch}")
        print(f"Model saved to {save_path}/MHRoberta_epoch_{epoch}")

    else:
        patience_counter += 1
        if patience_counter >= patience:
            print("Early stopping triggered")
            break

# 7️⃣ Loss & Accuracy Visualization
plt.figure(figsize=(10,6))
plt.plot(train_losses, label="Train Loss", marker="o")
plt.plot(val_losses, label="Validation Loss", marker="o")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training vs Validation Loss")
plt.legend()
plt.twinx()

plt.plot(val_accuracies, label="Validation Accuracy", color='green', linestyle="--", marker="s")
plt.ylabel("Accuracy")
plt.legend(loc="upper right")
plt.show()
```

```python

# ✅ Use PyTorch AMP (Automatic Mixed Precision)
scaler = torch.cuda.amp.GradScaler()

# ✅ Gradient Accumulation (for reducing memory usage)
gradient_accumulation_steps = 4  # Adjust as per available memory

# ✅ Early Stopping
best_val_loss = float("inf")
patience = 3
patience_counter = 0

# ✅ Track Losses
train_losses = []
val_losses = []

# ✅ Display Epoch Progress Bar
progress_bar = tqdm(range(num_epochs), desc="Epochs")  # 🛠️ Fix tqdm usage

for epoch in range(num_epochs):
    model.train()
    epoch_train_loss = 0.0

    for step, batch in enumerate(train_loader):
        batch = {k: v.to(device) for k, v in batch.items() if isinstance(v, torch.Tensor)}

        # ✅ Mixed Precision Training
        with autocast():
            outputs = model(**batch)
            loss = outputs.loss  # Do NOT manually divide loss

        # ✅ Backpropagation with Gradient Accumulation
        scaler.scale(loss).backward()

        if (step + 1) % gradient_accumulation_steps == 0:  # Update after N steps
            scaler.unscale_(optimizer)  # Unscale before clipping
            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)  # Clip gradients
            scaler.step(optimizer)
            scaler.update()
            optimizer.zero_grad()  # 🛠️ Move zero_grad() inside update condition
            scheduler.step()

        epoch_train_loss += loss.item()

    avg_train_loss = epoch_train_loss / len(train_loader)
    train_losses.append(avg_train_loss)

    # ✅ Validation Phase
    model.eval()
    epoch_val_loss = 0.0
    with torch.no_grad():
        for batch in val_loader:
            batch = {k: v.to(device) for k, v in batch.items() if isinstance(v, torch.Tensor)}
            outputs = model(**batch)
            epoch_val_loss += outputs.loss.item()

    avg_val_loss = epoch_val_loss / len(val_loader)
    val_losses.append(avg_val_loss)

    print(f"Epoch {epoch}: Avg Train Loss = {avg_train_loss:.4f} | Avg Val Loss = {avg_val_loss:.4f}")

    # ✅ Early Stopping & Model Saving
    if avg_val_loss < best_val_loss:
        best_val_loss = avg_val_loss
        patience_counter = 0
        model.save_pretrained(f"MHRoberta_epoch_{epoch}")
        tokenizer.save_pretrained(f"MHRoberta_epoch_{epoch}")
    else:
        patience_counter += 1
        if patience_counter >= patience:
            print("Early stopping triggered")
            break

    progress_bar.update(1)  # ✅ Update tqdm after each epoch

# ✅ Loss Visualization
plt.figure(figsize=(8,6))
plt.plot(range(len(train_losses)), train_losses, label="Train Loss", marker="o")
plt.plot(range(len(val_losses)), val_losses, label="Validation Loss", marker="o")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.title("Training vs Validation Loss")
plt.legend()
plt.show()
```

```python
import numpy as np
from sklearn.metrics import precision_recall_curve

def find_optimal_threshold(y_true, y_probs):
    precision, recall, thresholds = precision_recall_curve(y_true, y_probs)
    f1_scores = 2 * (precision * recall) / (precision + recall + 1e-9)
    best_idx = np.argmax(f1_scores)
    return thresholds[best_idx]

# After training, adjust threshold dynamically
y_probs = model.predict_proba(X_val)[:, 1]  # Get probability scores
optimal_threshold = find_optimal_threshold(y_val, y_probs)

# Use this threshold for final classification
final_preds = (y_probs > optimal_threshold).astype(int)
```

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    classification_report,
    confusion_matrix
)

# Set model to evaluation mode
model.eval()
all_predictions = []
all_labels = []

# Accumulate predictions and labels
with torch.no_grad():
    for batch in val_loader:
        batch = {k: v.to(device) for k, v in batch.items() if isinstance(v, torch.Tensor)}
        #batch = {k: v.to(device) for k, v in batch.items()}
        outputs = model(**batch)
        predictions = torch.argmax(outputs.logits, dim=1)
        all_predictions.extend(predictions.cpu().numpy())
        all_labels.extend(batch["labels"].cpu().numpy())

# Compute evaluation metrics
acc = accuracy_score(all_labels, all_predictions)
precision = precision_score(all_labels, all_predictions, average="macro")
recall = recall_score(all_labels, all_predictions, average="macro")
f1 = f1_score(all_labels, all_predictions, average="macro")

print(f"Accuracy: {acc:.4f}")
print(f"Precision (macro): {precision:.4f}")
print(f"Recall (macro): {recall:.4f}")
print(f"F1 Score (macro): {f1:.4f}")

# Detailed classification report and confusion matrix
report = classification_report(all_labels, all_predictions)
conf_matrix = confusion_matrix(all_labels, all_predictions)

print("\nClassification Report:\n", report)
print("\nConfusion Matrix:\n", conf_matrix)
```
